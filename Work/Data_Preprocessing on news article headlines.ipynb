```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
data=pd.read_csv('Headlines_2016.csv')
data
```

```python
data=data.iloc[:,1]
```

```python
data
```

```python
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
```

```python
data=data.str.lower()
```

```python
data_tokens=data.apply(word_tokenize)
```

```python
data_tokens
```

```python
tokens=[]
for head_line in data_tokens:
    tokens.extend(head_line)
tokens
data_tokens=tokens
```

```python
data_tokens
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stopwords_eng=stopwords.words('english')
stopwords_eng
```

```python
# data_tokens_ws=[]
for stopword in stopwords_eng:
    if stopword in data_tokens:
        for word in data_tokens:
            if word==stopword:
                data_tokens.remove(stopword)
#         print(data_tokens)
```
